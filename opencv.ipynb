```
import numpy as np
import cv2
import matplotlib.pyplot as plt
from urllib.request import urlopen
import pandas as pd
```

```

#fuction which converts url to image
def url_to_img(url):
  resp = urlopen(url)
  image = np.asarray(bytearray(resp.read()), dtype="uint8")
  image = cv2.imdecode(image, cv2.IMREAD_COLOR)
  return image
  
  
```

```
#horizontal
kernel = np.float32([[-1,2,-1],[-1,2,-1],[-1,2,-1]])
dst = cv2.filter2D(image1,-1,kernel)
cv2_imshow(dst)
plt.imshow(dst)
```
